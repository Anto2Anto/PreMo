```python
import pandas as pd
```

# Import raw data

```python
covid = pd.read_csv('covid_data.csv')
covid.describe()
```

# Reduce to "Austria" only

```python
austria=covid.iloc[12096:12900]
austria.reset_index(inplace=True)
austria=austria.fillna(0)
austria.describe()
```

# Correlation map

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
mask=np.zeros_like(austria.corr())
triangle_indices=np.triu_indices_from(mask)
mask[triangle_indices]=True
plt.figure(figsize=(25,20))
sns.heatmap(austria.corr(), mask=mask, annot=True, annot_kws={'size':18})
sns.set_style('white')
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.show()
```

# Drop unnecessary columns
## (where values are NaN or corr coef is close to 1)

```python
drop_list=['weekly_icu_admissions', 'weekly_icu_admissions_per_million',
       'weekly_hosp_admissions', 'weekly_hosp_admissions_per_million',
       'population', 'population_density', 'median_age', 'aged_65_older',
       'aged_70_older', 'gdp_per_capita', 'extreme_poverty',
       'cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers',
       'male_smokers', 'handwashing_facilities', 'hospital_beds_per_thousand',
       'life_expectancy', 'human_development_index','new_cases_smoothed','new_deaths_smoothed',
     'new_cases_smoothed_per_million', 'new_deaths_smoothed_per_million', 'new_tests_smoothed_per_thousand',
     'new_vaccinations_smoothed_per_million', 'new_people_vaccinated_smoothed',
       'new_people_vaccinated_smoothed_per_hundred', 'total_cases_per_million',
       'new_cases_per_million', 'total_deaths_per_million', 'new_deaths_per_million', 'new_vaccinations_smoothed',
       'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred',
       'people_fully_vaccinated_per_hundred', 'total_boosters_per_hundred', 'excess_mortality_cumulative_per_million',
     'icu_patients_per_million', 'hosp_patients_per_million', 'total_tests_per_thousand', 'new_tests_per_thousand']
austria.drop(columns=drop_list, inplace=True)
```

# New correlation map

```python
mask=np.zeros_like(austria.corr())
triangle_indices=np.triu_indices_from(mask)
mask[triangle_indices]=True
plt.figure(figsize=(25,20))
ax=sns.heatmap(austria.corr(), mask=mask, annot=True, annot_kws={'size':16})
sns.set_style('white')
cbar=ax.collections[0].colorbar
cbar.ax.tick_params(labelsize=20)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)
plt.show()
```

```python
# vaccinated - cases, deaths
new_vacs=austria['new_vaccinations'].tolist()
new_cases=austria['new_cases'].tolist()
new_deaths=austria['new_deaths'].tolist()
date=austria['date'].tolist()
X=austria.index
```

# Polynomial Regression

```python
from sklearn import metrics
```

```python
p1=np.poly1d(np.polyfit(X, new_vacs, 45))
p2=np.poly1d(np.polyfit(X, new_cases, 35))
p3=np.poly1d(np.polyfit(X, new_deaths, 35))
P1=np.round(metrics.r2_score(new_vacs, p1(X)),3)
P2=np.round(metrics.r2_score(new_cases, p2(X)),3)
P3=np.round(metrics.r2_score(new_deaths, p3(X)),3)
print(f'r2_score for p1={P1}')
print(f'r2_score for p2={P2}')
print(f'r2_score for p3={P3}')
```

# Random Forest

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
x=austria[['index']]
```

```python
vac_forest_model = RandomForestRegressor(max_depth=5, n_estimators=5, random_state=1)
vac_forest_model.fit(x, new_vacs)
y_rf_vac = vac_forest_model.predict(x)
#-----------------------------
case_forest_model = RandomForestRegressor(max_depth=5, n_estimators=5, random_state=1)
case_forest_model.fit(x, new_cases)
y_rf_case = case_forest_model.predict(x)
#-----------------------------
death_forest_model = RandomForestRegressor(max_depth=5, n_estimators=5, random_state=1)
death_forest_model.fit(x, new_deaths)
y_rf_death = death_forest_model.predict(x)
```

```python
fig=plt.figure(figsize=(20,20))
gs=gridspec.GridSpec(3,3)
#-------------------------------
ax1=fig.add_subplot(gs[0,:])
ax1.scatter(date, new_vacs, label='new vaccinations', marker='.', c='blue')
ax1.scatter(date, new_cases, label='new cases', marker='.', c='orange')
ax1.scatter(date, new_deaths, label='new deaths', marker='.', c='green')
ax1.set_yscale('log')
ax1.xaxis.set_tick_params(rotation=0, direction='out', length=5, labelsize=15)
ax1.yaxis.set_tick_params(labelsize=15)
ax1.xaxis.set_major_locator(MultipleLocator(180))
ax1.xaxis.set_minor_locator(MultipleLocator(30))
ax1.grid(which='major', linewidth=2, color='grey')
ax1.grid(which='minor')
ax1.legend(loc='upper left',fontsize='20', shadow=True)
ax1.set_title('New cases, deaths and vaccinations over a period of last 3 years', fontsize='25', fontstyle='italic')
#------------------------------
ax2=fig.add_subplot(gs[1,0])
ax2.scatter(date, new_vacs, marker='.', c='blue')
ax2.plot(date, p1(X), c='r', linewidth=3, label='Poly')
ax2.plot(date, y_rf_vac, c='black', linewidth=3, label='RF')
ax2.set_xlim(xmin='2020-12-21')
ax2.xaxis.set_tick_params(rotation=0, direction='out', length=5, labelsize=15)
ax2.yaxis.set_tick_params(labelsize=15)
ax2.xaxis.set_major_locator(MultipleLocator(180))
ax2.xaxis.set_minor_locator(MultipleLocator(30))
ax2.grid(which='major', linewidth=2, color='grey')
ax2.grid(which='minor')
ax2.legend(loc='upper left',fontsize='20', shadow=True)
ax2.set_title('New vaccinations only', fontsize='25', fontstyle='italic')
#------------------------------
ax3=fig.add_subplot(gs[1,1])
ax3.scatter(date, new_cases, marker='.', c='orange')
ax3.plot(date, p2(X), c='r', linewidth=3, label='Poly')
ax3.plot(date, y_rf_case, c='black', linewidth=2, label='RF')
ax3.xaxis.set_tick_params(rotation=0, direction='out', length=5, labelsize=15)
ax3.yaxis.set_tick_params(labelsize=15)
ax3.xaxis.set_major_locator(MultipleLocator(360))
ax3.xaxis.set_minor_locator(MultipleLocator(30))
ax3.grid(which='major', linewidth=1, color='grey')
ax3.grid(which='minor')
ax3.legend(loc='upper left',fontsize='20', shadow=True)
ax3.set_title('New cases only', fontsize='25', fontstyle='italic')
#------------------------------
ax4=fig.add_subplot(gs[1,2])
ax4.scatter(date, new_deaths, marker='.', c='green')
ax4.plot(date, p3(X), c='r', linewidth=3, label='Poly')
ax4.plot(date, y_rf_death, c='black', linewidth=3, label='RF')
ax4.xaxis.set_tick_params(rotation=0, direction='out', length=5, labelsize=15)
ax4.yaxis.set_tick_params(labelsize=15)
ax4.xaxis.set_major_locator(MultipleLocator(360))
ax4.xaxis.set_minor_locator(MultipleLocator(30))
ax4.grid(which='major', linewidth=1, color='grey')
ax4.grid(which='minor')
ax4.legend(loc='upper right',fontsize='20', shadow=True)
ax4.set_title('New deaths only', fontsize='25', fontstyle='italic')
#------------------------------
plt.show()
```

# Conclusions and insights:

#### 1. The first figure clearly illustrates the dependancies between number of new vaccinations, cases and deaths.
#### 2. The blue line starts in Feb 2021. Apparently Austria got first doses of vaccines that month.
#### 3. After noticeble increasing of vaccinations between Feb and May 2021, cases and deaths start to decrease since that very May. But this delay of 3 months (Feb to May) imho might be interpreted as reaching balance between covid spread and vacc. growth. Because getting few people vaccinated isn't enough to stop the disease.
#### 4. Just after the number of new vaccinations stopped increasing near June 2021, number of cases and deaths immediately started growing up. Then the whole situation repeats al least once.
#### 5. The green line carefully repeats the form of the orange one. The correlation coefficient is about 0.155, hence the number of new deaths is way lower than the number of new cases. It may be a proof, that COVID-19 isn't a humicide desease or a new pestilence, but it's still extremely dangerous.
#### 6. Second, third and forth figures show all three sets separately. Polynomial Regression (PR) and Random Forest (RF) were perfomed here as predictive models. I had to choose highest degrees for PR (35-45) to gain good prediction accuracy. Whilst few number of more or less shallow decision trees (max_depth=5, n_estimators=5) was completely enough to build a really good model with better precision.
